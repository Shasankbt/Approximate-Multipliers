```python
from torch.utils.cpp_extension import load
import torch

# 1. Compile
# Make sure the file above is saved as "pbom8-73y/pbo_float.cpp"
pbo_cpp = load(
    name="pbo_extension",
    sources=[
        "pbom8-73y/multiplier.cpp",
        "pbom8-73y/matmul.cpp"
    ],
    verbose=True,
    extra_cflags=["-O3"]
)
# 2. Test Data (2D Tensors)
a = torch.tensor([[[10.0]]], dtype=torch.float16)
b = torch.tensor([[[4.0]]], dtype=torch.float16)

# 3. Run
print("Exact Math:\n", a.matmul(b))

approx_output = pbo_cpp.tensor_product(a, b)
print("\nApproximate BFloat16 Math:\n", approx_output)
```
